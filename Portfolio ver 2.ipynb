```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('http://taanila.fi/data1.xlsx', sheet_name='Data')
df.head()
```

```python
df.columns
```

```python
df.rename(columns = {'sukup':'sukupuoli','palveluv':'palveluvuodet','työtov':'työtoverit','työymp':'työympäristö','palkkat':'palkkkataso',
                    'työteht':'työtehtävä','työterv':'työterveyshuolto','lomaosa':'lomaosake','kuntosa':'kuntosali'}, inplace=True)
df.head()
```

```python
df['sukupuoli']=df['sukupuoli'].replace({1 : 'Mies', 2 : 'Nainen'})
df['perhe']=df['perhe'].replace({1 : 'Perheellinen', 2 : 'Perheetön'})
df.head()
```

```python
bins= [0, 5, 10, 15, 20, 25]
df['Palvelusvuodet']=pd.cut(df['palveluvuodet'], bins = bins)
df.head()
```

```python
df[df['ikä']<25]
```

```python
df[['sukupuoli','palkka']].sort_values(by='palkka', ascending=False)
df.head(20)
```

```python
df.drop(['johto','työympäristö','työtehtävä','lomaosake','hieroja','kuntosali'], axis=1).head(10)
```

```python
koulutus = ['Peruskoulu', '2. Aste', 'Korkeakoulu', 'Ylempi korkeakoulu']
perhe = ['Perheellinen', 'Perheetön']
sukupuoli = ['Mies', 'Nainen']
tyytyväisyys = ['Erittäin tyytymätön', ' Jokseenkin tyytymätön', 'Ei tyytymätön eikä tyytyväinen',
               'Jokseenkin tyytyväinen', 'Erittäin tyytyväinen']
```

```python
df1 = pd.crosstab(df['koulutus'], 'n')
df1.index = koulutus
df1.columns.name = ''
df1
```

```python
df1['%'] = df1/df1.sum()
df1.style.format({'%': '{:.1%}'})
```

```python
ax = df1['%'].plot.barh(color = 'C3')
vals = ax.get_xticks()
ax.set_xticklabels(['{:.0%}'.format(x) for x in vals])
```
