```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

sns.set()
```

```python
wine=pd.read_excel('http://taanila.fi/wine.xlsx')
wine.head(25)
```

```python
X = wine.drop('Y', axis=1)
```

```python
from sklearn.cluster import KMeans

malli = KMeans(n_clusters=3)
malli.fit(X)

malli.cluster_centers_
```

```python
X['K'] = malli.predict(X)
pd.crosstab(X['K'], 'lkm')
```

```python
X.groupby('K').describe().T
```

```python
pd.crosstab(X['K'], wine['Y'])
```

```python
sns.pairplot(X, hue='K')
```
