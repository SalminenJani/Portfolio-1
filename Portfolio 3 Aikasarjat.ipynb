```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
sampo=pd.read_excel('sampo_14_19.xlsx')
sampo.head()
```

```python
nordea=pd.read_excel('nordea_14_19.xlsx')
nordea.head()
```

```python
sampo.index=pd.to_datetime(sampo['Date'])
sampo.sort_index(inplace=True)
sampo.head()
```

```python
nordea.index=pd.to_datetime(nordea['Date'])
nordea.sort_index(inplace=True)
nordea.head()
```

```python
sampo['Average price'].plot()
```

```python
nordea['Average price'].plot()
```

```python
sampo['Turnover']['2015'].plot()
```

```python
nordea['Turnover']['2015'].plot()
```

```python
fig1,ax1=plt.subplots(figsize = (10,7))
color='C0'
ax1.set_ylabel('Nordea', color=color)
ax1.plot(nordea['Turnover']['2015'], color=color)

ax2=ax1.twinx()

color='C1'
ax2.set_ylabel('Sampo', color=color)
ax2.plot(sampo['Turnover']['2015'], color=color)
```

```python
fig2,ax3 = plt.subplots(figsize = (10,7))

color='C2'
ax3.set_ylabel('Nordea', color=color)
ax3.plot(nordea['High price']['2017-01'], color=color)
ax3.tick_params(axis='y', labelcolor='C4')

ax4=ax3.twinx()

color='C3'
ax4.set_ylabel('Sampo', color=color)
ax4.plot(sampo['High price']['2017-01'], color=color)
ax4.tick_params(axis='y', labelcolor='C5')
```

```python
fig3,ax5 = plt.subplots(figsize= (10,7))

color='C2'
ax5.set_ylabel('High price', color=color)
ax5.plot(nordea['High price']['2017-05'], color=color)

ax6=ax5.twinx()

color='C3'
ax6.set_ylabel('Low price', color=color)
ax6.plot(nordea['Low price']['2017-05'], color=color)
```

```python
nordea['Average price'].plot()

nordea['Average price'].rolling(100).mean().plot()
```

```python
nordea['Nordea change']=nordea['Closing price'].pct_change()
nordea.head()
```

```python
sampo['Sampo change']=sampo['Closing price'].pct_change()
sampo.head()
```

```python
muutokset=pd.concat([nordea['Nordea change'], sampo['Sampo change']], axis=1)
muutokset.head()
```

```python
ax7=muutokset['2019-01':].plot()
ax7.set_ylabel('Muutos')
vals=ax7.get_yticks()
ax7.set_yticklabels(['{:.0f} %'.format(y*100) for y in vals])
```

```python
muutokset.describe()
```

```python
muutokset[(muutokset['Nordea change']<-0.05)|(muutokset['Sampo change']>0.05)]
```

```python
muutokset.corr()
```

```python
korrelaatio=muutokset['Nordea change'].rolling(150,min_periods=30).corr(muutokset['Sampo change'])
korrelaatio.plot()
```

```python
muutokset['Weekday']=muutokset.index.weekday
muutokset.head()
```

```python
muutokset.groupby('Weekday')['Nordea change'].describe()
```

```python
muutokset.groupby('Weekday')['Sampo change'].describe()
```
